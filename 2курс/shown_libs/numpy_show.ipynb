**Агрегаторы и простые фильтры**

```python
print("A", A)
print('min\n', np.min(A), '\n')
print('max\n', np.max(A, 0), '\n')
print('mean\n', np.mean(A, 0), '\n')
print('sum\n', np.sum(A, 1), '\n')
```

```python
print(A.ravel())
print(np.mean(A.ravel()))
print(np.median(A.ravel()))
print(np.sum(A.ravel())/A.ravel().size)
```

```python
m = A.ravel()
m[-1] = 1000 
print(m.ravel())
print(np.mean(m.ravel()))
print(np.median(m.ravel()))
```
