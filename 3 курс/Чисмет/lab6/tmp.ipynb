```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import mark_inset, zoomed_inset_axes
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x) + 0.1 * np.random.randn(100)

# Main plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Overview with zoomed insets")

# First zoomed-in region
axins1 = zoomed_inset_axes(ax, zoom=2, loc='lower left')  # zoom factor, position
axins1.plot(x, y)
axins1.set_xlim(1, 2)  # region to zoom
axins1.relim()
axins1.autoscale_view()
mark_inset(ax, axins1, loc1=2, loc2=4, fc="none", ec="0.5")

# Second zoomed-in region
axins2 = zoomed_inset_axes(ax, zoom=4, loc='lower right')
axins2.plot(x, y)
axins2.set_xlim(7, 8)
axins2.set_ylim(np.min(y[(x >= 7) & (x <= 8)]), np.max(y[(x >= 7) & (x <= 8)]))
mark_inset(ax, axins2, loc1=2, loc2=4, fc="none", ec="0.5")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)


fig = plt.figure(figsize=(8, 4))  # Create a figure of specific size

# Add first axes (plot)
ax1 = fig.add_subplot(1, 2, 1)  # 1 row, 2 columns, first plot
ax1.plot(x, y)
ax1.set_title("Sine")

# Add second axes (plot)
ax2 = fig.add_subplot(1, 2, 2)  # second plot
ax2.plot(x, y)
ax2.set_xlim(1, 2)

ax2.autoscale_view()
ax2.set_title("Cosine")

fig.suptitle("Trigonometric Functions")  # Title for the whole figure
plt.tight_layout()
plt.show()
```
